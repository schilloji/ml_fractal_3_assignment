```python
from google.colab import drive
import numpy as np
import pandas as pd
from PIL import Image
from IPython.display import display
import os
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.optimizers import Adam

drive.mount('/content/drive')
```

```python
def dataLoader(path):
  # set the image size for resizing
  img_size = (28, 28)
  num_classes = 10

  # create empty arrays for the images and labels
  images = []
  labels = []

  # loop through each folder in the data path
  for folder in os.listdir(data_path):
      folder_path = os.path.join(data_path, folder)
      
      # loop through each image in the folder
      for file in os.listdir(folder_path):
          file_path = os.path.join(folder_path, file)
          
          # open the image using Pillow
          img = Image.open(file_path)
          
          # convert the image to RGB mode if it's not already in RGB
          if img.mode != 'RGB':
              img = img.convert('RGB')
          
          # resize the image to the desired size
          img = img.resize(img_size)
          
          # convert the image to a NumPy array
          img_array = np.array(img,dtype=np.float32)
          img_array=np.ravel(img_array)
          img_array /= 255.0
          
          # add the image and label to the respective arrays
          images.append(img_array)
          labels.append(folder)
          
  # convert the images and labels to NumPy arrays
  images = np.array(images)
  labels = np.array(labels,dtype=np.int)
  labels = np.eye(num_classes)[labels]

  # print the shapes of the arrays
  print('Images shape:', images.shape)
  print('Labels shape:', labels.shape)

  return images,labels
```

```python
# set the path to the directory containing images
data_path = '/content/drive/MyDrive/GurNum/train/'
train_images,train_labels=dataLoader(data_path)

data_path = '/content/drive/MyDrive/GurNum/val/'
val_images,val_labels=dataLoader(data_path)
```

```python
# Define the network architecture
model = Sequential()
model.add(Flatten(input_shape=(2352,)))  # Input layer
model.add(Dense(128, activation='relu'))  # Hidden layer 1
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(64, activation='relu'))  # Hidden layer 2
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(10, activation='softmax'))  # Output layer with 10 classes

# Compile the model with an appropriate loss function and optimizer
model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.001), metrics=['accuracy'])

# Train the model on the training set and validate it on the validation set
model.fit(train_images, train_labels, epochs=200, batch_size=128,verbose=1)

# Evaluate the model on the test set
loss, accuracy = model.evaluate(val_images, val_labels)
print('Val loss:', loss)
print('Val accuracy:', accuracy)

```
